```python
# Import Libraries
from google_play_scraper import Sort, reviews
import pandas as pd
import numpy as np
```

```python
# Scrape 100000 reviews
result, continuation_token = reviews(
    'com.bibit.bibitid',           # App ID for Shopee Indonesia
    lang='id',                 # Language (in Indonesian)
    country='id',              # Country (Indonesia)
    sort=Sort.NEWEST,         # Sort by newest reviews
    count=100000,               # Get 100000 reviews
    filter_score_with=None     # Include all scores (1-5)
)

# Create DataFrame from the result
df_busu = pd.DataFrame(result)

# Display the first few rows of the DataFrame
print(df_busu.head())

# Show the info
df_busu.info()
```

```python
#convert df_busu to excel file
df_busu.to_csv('df_busu.csv', index=False)
```

```python
#Convert df_busu to csv file
df_busu.to_csv('df_busu.csv', index=False)
```
